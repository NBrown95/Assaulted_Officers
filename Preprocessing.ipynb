```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
from config import db_password
```

```python
file_path = "AssaultedOfficerData.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
# Remove data before 2019
df['DATE_OCCURED'] = pd.to_datetime(df['DATE_OCCURED'])
df
```

```python
# Get current data
current_df = df[~(df['DATE_OCCURED'] < '2019-01-01')]
current_df.reset_index
```

```python
# Drop columns
current_df = current_df.drop(columns=['ID', 'DATE_REPORTED', 'NIBRS_CODE', 'UCR_HIERARCHY', 'LMPD_DIVISION', 'LMPD_BEAT', 'BLOCK_ADDRESS', 'LEOKA_APPLIES', 'OFFICER_KILLED', 'OFFENSE_CITY', 'OFFENSE_ZIP', 'ATT_COMP'])
```

```python
# Check datatypes
current_df.info()
```

```python
# Check dataframe
current_df.head()
```

```python
# Begin preprocessing of crime dataset
file_path2 = "Crime_Data_2019.csv"
crime_df = pd.read_csv(file_path2)
crime_df.head()
```

```python
# Change datatype to datetime
crime_df['DATE_OCCURED'] = pd.to_datetime(crime_df['DATE_OCCURED'])
crime_df
```

```python
# Only focus on assaults
crime_df = crime_df[crime_df['CRIME_TYPE']=='ASSAULT']
```

```python
# Drop irrelevant columns
clean_crime_df = crime_df.drop(columns=['ID', 'DATE_REPORTED', 'NIBRS_CODE', 'UCR_HIERARCHY', 'LMPD_DIVISION', 'LMPD_BEAT', 'BLOCK_ADDRESS', 'ATT_COMP', 'CITY', 'ZIP_CODE'])
clean_crime_df
```

```python
# Drop null values
clean_crime_df.dropna()
```

```python
# Check final datatypes
clean_crime_df.info()
```

```python
# Create connection string
db_string = f"postgres://postgres:{db_password}@127.0.0.1:5432/Assaulted_Officers"
```

```python
# Create Database engine
engine = create_engine(db_string)
```

```python
# import data to sql
current_df.to_sql(name='Assaults_', con=engine)
clean_crime_df.to_sql(name='Crime_', con=engine)
```
